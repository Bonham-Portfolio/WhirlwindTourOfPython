# Loops

## Loops - Notes

for loops iterate over a sequence. 

A range can be defined for for loops

while loops loop until a criteria is met

break exits a loop

continue can skip a block and reloop until criteria is met

else is an alternative to the for loop. a break can skip an else clause



```
evens = [2, 4, 6, 8, 10,]
for even in evens:
    print(even)
print("\n")

print("for loop examples in ranges")
for x in range(6):
    print(x)
for x in range(5, 9):
    print(x)
for x in range(4, 10, 2):
    print(x)
print("\n")

print("While loop using a limit")
twos = 0
while twos < 10:
    print(twos)
    twos += 2 
print("\n")

print("breaking a loop")
twos = 0
while True:
    print(twos)
    twos += 2
    if twos >= 10:
        break


print("using a continue statement to loop back to for")
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)

print("Else example")
limit = 0
while(limit<5):
    print(limit)
    limit +=1
else:
    print("The limit of %d has been reached" %(limit))







```

# Classes and Objects

## Classes and Objects

Classes are a collection of variables and functions. 

You can build a class that has multiple variables and functions then call it and reuse that information as a template

You can create variable objects to house classes. You can make changes to variables within a class using objects and the changes will be independent of each other.

```
class classified:
    asu = "Kyle is the best"

    def function(self):
        print("Kyle will get an A!")

no_change = classified()
changed = classified()

changed.asu = "Kyle will get an A+"

print(no_change.asu)
print(changed.asu)

print(no_change.function())
```
